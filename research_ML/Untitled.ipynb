```python
from rdkit import Chem
from rdkit.Chem import AllChem

# Define the SMILES string
smiles = 'CCCCCCCCCC'

# Convert the SMILES string to a RDKit molecule object
mol = Chem.MolFromSmiles(smiles)

# Calculate the Morgan fingerprint using the default parameters (radius=2, nBits=2048)
fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=3, nBits=2048)

# Convert the fingerprint to a numerical vector representation
fp_vector = fp.ToBitString()
print(fp_vector)
```
